```python
# --- Imports ---
import pandas as pd
import requests
from bs4 import BeautifulSoup
import feedparser
import plotly.graph_objects as go

# --- Load CSV ---
# Replace with your Drive path if using Google Drive
data = pd.read_csv("momentum_data.csv")

# --- Scraper Functions ---

def scrape_github_repos(query, max_repos=10):
    url = f"https://github.com/search?q={query}&type=repositories"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    repos = [a.text.strip() for a in soup.select("a.v-align-middle")][:max_repos]
    return repos

def scrape_product_hunt(category, max_items=10):
    url = f"https://www.producthunt.com/topics/{category}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    products = [a.text.strip() for a in soup.select("h3")][:max_items]
    return products

def scrape_hn_stories(query, max_items=10):
    url = f"https://hn.algolia.com/api/v1/search?query={query}"
    response = requests.get(url).json()
    stories = [hit["title"] for hit in response["hits"]][:max_items]
    return stories

def scrape_rss_feed(url, max_items=10):
    feed = feedparser.parse(url)
    return [entry.title for entry in feed.entries[:max_items]]

def scrape_x_startups(query, max_items=10):
    url = f"https://twitter.com/search?q={query}&f=live"
    headers = {"User-Agent": "Mozilla/5.0"}
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")
        tweets = [t.get_text(strip=True) for t in soup.find_all("span")]
        startups = tweets[:max_items]
        return startups
    except Exception as e:
        print(f"X scraper error: {e}")
        return []

# --- Aggregate new startups ---
new_startups = list(set(
    scrape_github_repos("AI startup") +
    scrape_product_hunt("ai") +
    scrape_hn_stories("AI startup") +
    scrape_rss_feed("https://techcrunch.com/feed/") +
    scrape_x_startups("AI startup")
))
print(f"New startups detected: {new_startups}")

# --- Leader Detection ---
def detect_leaders(data):
    today_leader = None
    today_score = -1
    week_leader = None
    week_score = -1
    jump_leader = None
    jump_score = -999

    for index, row in data.iterrows():
        trajectory = eval(row["Trajectory"])
        latest = trajectory[-1]
        weekly_total = sum(trajectory)
        jump = trajectory[-1] - trajectory[-2] if len(trajectory) >= 2 else 0

        if latest > today_score:
            today_score = latest
            today_leader = row["Startup"]

        if weekly_total > week_score:
            week_score = weekly_total
            week_leader = row["Startup"]

        if jump > jump_score:
            jump_score = jump
            jump_leader = row["Startup"]

    return {
        "today_leader": today_leader,
        "today_score": today_score,
        "week_leader": week_leader,
        "week_score": week_score,
        "jump_leader": jump_leader,
        "jump_score": jump_score
    }

# --- Plot Chart with Mini Leaderboard ---
def plot_top_momentum_chart_with_leaderboard(data, top_n=10, new_startups=None):
    if new_startups is None:
        new_startups = []

    leaders = detect_leaders(data)
    top_leader = leaders["today_leader"]
    week_leader = leaders["week_leader"]
    jump_leader = leaders["jump_leader"]

    leaderboard_text = f"""
    🥇 Today’s Leader: {top_leader} ({leaders['today_score']})<br>
    📈 Weekly Leader: {week_leader} ({leaders['week_score']})<br>
    🚀 Biggest Jump: {jump_leader} ({leaders['jump_score']})
    """

    fig = go.Figure()
    fig.add_annotation(
        xref="paper", yref="paper",
        x=0.5, y=1.08,
        text=leaderboard_text,
        showarrow=False,
        font=dict(size=14, color="black"),
        align="center",
        bgcolor="lightgray",
        bordercolor="black",
        borderwidth=1,
        borderpad=4
    )

    latest_scores = {row["Startup"]: eval(row["Trajectory"])[-1] for idx, row in data.iterrows()}
    top_companies = sorted(latest_scores, key=latest_scores.get, reverse=True)[:top_n]
    display_companies = set(top_companies + new_startups)

    for idx, row in data.iterrows():
        startup = row["Startup"]
        if startup not in display_companies:
            continue

        trajectory = eval(row["Trajectory"])
        for i in range(1, len(trajectory)):
            y0 = trajectory[i-1]
            y1 = trajectory[i]
            x0 = i
            x1 = i+1

            color = "green" if y1 > y0 else "red" if y1 < y0 else "gray"
            width = 3
            if startup == top_leader:
                color = "gold"
                width = 6
            elif startup in new_startups:
                color = "blue"

            fig.add_trace(go.Scatter(
                x=[x0, x1],
                y=[y0, y1],
                mode="lines+markers",
                line=dict(color=color, width=width),
                name=startup if i == 1 else None,
                legendgroup=startup,
                showlegend=(i == 1)
            ))

    fig.update_layout(
        title=f"Momentum Intelligence – Top {top_n} + New Startups",
        xaxis_title="Check-in Number",
        yaxis_title="Momentum Score",
        yaxis=dict(range=[0, 100]),
        hovermode="closest"
    )
    fig.show()

# --- Run the chart ---
plot_top_momentum_chart_with_leaderboard(data, top_n=10, new_startups=new_startups)
```

```python
import random
from datetime import datetime

# Function to generate a new momentum score (simulate real data)
def get_new_momentum(prev_score=None):
    if prev_score is None:
        return random.randint(1, 20)  # starting score for new startup
    # small variation around last score
    change = random.randint(-5, 10)
    return max(prev_score + change, 0)  # ensure score >= 0

# Load current startups
data = pd.read_sql_query("SELECT * FROM startups", conn)

# Update trajectories
for idx, row in data.iterrows():
    name = row['name']
    trajectory = eval(row['trajectory'])  # convert string to list
    new_score = get_new_momentum(trajectory[-1])
    trajectory.append(new_score)

    # Update DB
    cursor.execute('''
    UPDATE startups
    SET trajectory = ?, last_checkin = ?
    WHERE name = ?
    ''', (str(trajectory), datetime.now(), name))

# Commit updates
conn.commit()

# Reload updated data
data = pd.read_sql_query("SELECT * FROM startups", conn)

print("Trajectories updated! Here's a preview:")
data.head()
```

```python
# --- Imports ---
import pandas as pd
import requests
from bs4 import BeautifulSoup
import feedparser
import plotly.graph_objects as go
import sqlite3

# --- Connect to SQLite DB ---
conn = sqlite3.connect("live.db")
cursor = conn.cursor()

# --- Load startups from SQLite ---
data = pd.read_sql_query("SELECT * FROM startups", conn)

# --- Scraper Functions ---
def scrape_github_repos(query, max_repos=10):
    url = f"https://github.com/search?q={query}&type=repositories"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    repos = [a.text.strip() for a in soup.select("a.v-align-middle")][:max_repos]
    return repos

def scrape_product_hunt(category, max_items=10):
    url = f"https://www.producthunt.com/topics/{category}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    products = [a.text.strip() for a in soup.select("h3")][:max_items]
    return products

def scrape_hn_stories(query, max_items=10):
    url = f"https://hn.algolia.com/api/v1/search?query={query}"
    response = requests.get(url).json()
    stories = [hit["title"] for hit in response["hits"]][:max_items]
    return stories

def scrape_rss_feed(url, max_items=10):
    feed = feedparser.parse(url)
    return [entry.title for entry in feed.entries[:max_items]]

def scrape_x_startups(query, max_items=10):
    url = f"https://twitter.com/search?q={query}&f=live"
    headers = {"User-Agent": "Mozilla/5.0"}
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")
        tweets = [t.get_text(strip=True) for t in soup.find_all("span")]
        startups = tweets[:max_items]
        return startups
    except Exception as e:
        print(f"X scraper error: {e}")
        return []

# --- Run scrapers ---
github_startups = scrape_github_repos("AI startup", max_repos=10)
ph_startups = scrape_product_hunt("ai", max_items=10)
hn_startups = scrape_hn_stories("AI startup", max_items=10)
rss_startups = scrape_rss_feed("https://techcrunch.com/feed/", max_items=10)
x_startups = scrape_x_startups("AI startup", max_items=10)

# --- Safe merge ---
existing_startups = data['name'].tolist()
new_startups = list(set(existing_startups + github_startups + ph_startups + hn_startups + rss_startups + x_startups))

# --- Insert new startups into DB ---
for s in new_startups:
    if s not in existing_startups:
        cursor.execute('''
        INSERT INTO startups (name, trajectory)
        VALUES (?, ?)
        ''', (s, '[0]'))  # initialize trajectory with [0]

conn.commit()

# --- Reload data from DB after insertion ---
data = pd.read_sql_query("SELECT * FROM startups", conn)

# --- Leader Detection ---
def detect_leaders(data):
    today_leader = None
    today_score = -1
    week_leader = None
    week_score = -1
    jump_leader = None
    jump_score = -999

    for index, row in data.iterrows():
        trajectory = eval(row["trajectory"])
        latest = trajectory[-1]
        weekly_total = sum(trajectory)
        jump = trajectory[-1] - trajectory[-2] if len(trajectory) >= 2 else 0

        if latest > today_score:
            today_score = latest
            today_leader = row["name"]

        if weekly_total > week_score:
            week_score = weekly_total
            week_leader = row["name"]

        if jump > jump_score:
            jump_score = jump
            jump_leader = row["name"]

    return {
        "today_leader": today_leader,
        "today_score": today_score,
        "week_leader": week_leader,
        "week_score": week_score,
        "jump_leader": jump_leader,
        "jump_score": jump_score
    }

# --- Plot Chart with Mini Leaderboard ---
def plot_top_momentum_chart_with_leaderboard(data, top_n=10, new_startups=None):
    if new_startups is None:
        new_startups = []

    leaders = detect_leaders(data)
    top_leader = leaders["today_leader"]
    week_leader = leaders["week_leader"]
    jump_leader = leaders["jump_leader"]

    leaderboard_text = f"""
    🥇 Today’s Leader: {top_leader} ({leaders['today_score']})<br>
    📈 Weekly Leader: {week_leader} ({leaders['week_score']})<br>
    🚀 Biggest Jump: {jump_leader} ({leaders['jump_score']})
    """

    fig = go.Figure()
    fig.add_annotation(
        xref="paper", yref="paper",
        x=0.5, y=1.08,
        text=leaderboard_text,
        showarrow=False,
        font=dict(size=14, color="black"),
        align="center",
        bgcolor="lightgray",
        bordercolor="black",
        borderwidth=1,
        borderpad=4
    )

    latest_scores = {row["name"]: eval(row["trajectory"])[-1] for idx, row in data.iterrows()}
    top_companies = sorted(latest_scores, key=latest_scores.get, reverse=True)[:top_n]
    display_companies = set(top_companies + new_startups)

    for idx, row in data.iterrows():
        startup = row["name"]
        if startup not in display_companies:
            continue

        trajectory = eval(row["trajectory"])
        for i in range(1, len(trajectory)):
            y0 = trajectory[i-1]
            y1 = trajectory[i]
            x0 = i
            x1 = i+1

            color = "green" if y1 > y0 else "red" if y1 < y0 else "gray"
            width = 3
            if startup == top_leader:
                color = "gold"
                width = 6
            elif startup in new_startups:
                color = "blue"

            fig.add_trace(go.Scatter(
                x=[x0, x1],
                y=[y0, y1],
                mode="lines+markers",
                line=dict(color=color, width=width),
                name=startup if i == 1 else None,
                legendgroup=startup,
                showlegend=(i == 1)
            ))

    fig.update_layout(
        title=f"Momentum Intelligence – Top {top_n} + New Startups",
        xaxis_title="Check-in Number",
        yaxis_title="Momentum Score",
        yaxis=dict(range=[0, 100]),
        hovermode="closest"
    )
    fig.show()

# --- Run the chart ---
plot_top_momentum_chart_with_leaderboard(data, top_n=10, new_startups=new_startups)
```

```python
import plotly.graph_objects as go
from IPython.display import display, HTML

# --- Leader Detection ---
def detect_leaders(data):
    today_leader = None
    today_score = -1

    weekly_leader = None
    weekly_score = -1

    biggest_jump = None
    biggest_jump_score = -999

    for _, row in data.iterrows():
        trajectory = eval(row["trajectory"])

        latest = trajectory[-1]
        weekly = sum(trajectory)
        jump = trajectory[-1] - trajectory[-2] if len(trajectory) > 1 else 0

        if latest > today_score:
            today_score = latest
            today_leader = row["name"]

        if weekly > weekly_score:
            weekly_score = weekly
            weekly_leader = row["name"]

        if jump > biggest_jump_score:
            biggest_jump_score = jump
            biggest_jump = row["name"]

    return today_leader, today_score, weekly_leader, weekly_score, biggest_jump, biggest_jump_score


today_leader, today_score, weekly_leader, weekly_score, biggest_jump, biggest_jump_score = detect_leaders(data)


# --- Leaderboard UI ---
display(HTML(f"""
<h1 style='font-size:38px'>Momentum Intelligence</h1>

<div style="display:flex;gap:40px;font-size:20px">

<div>
<b>Today's Leader</b><br>
{today_leader} ({today_score})
</div>

<div>
<b>Weekly Leader</b><br>
{weekly_leader} ({weekly_score})
</div>

<div>
<b>Biggest Jump</b><br>
{biggest_jump} (+{biggest_jump_score})
</div>

</div>
"""))


# --- Chart ---
fig = go.Figure()

for _, row in data.iterrows():

    trajectory = eval(row["trajectory"])

    fig.add_trace(
        go.Scatter(
            y=trajectory,
            mode="lines",
            name=row["name"],
            line=dict(width=2)
        )
    )

fig.update_layout(
    height=600,
    template="plotly_dark",
    title="Startup Momentum Field"
)

fig.show()
```

```python
import pandas as pd
from IPython.display import display

leaderboard = []

for _, row in data.iterrows():

    trajectory = eval(row["trajectory"])

    latest = trajectory[-1]
    prev = trajectory[-2] if len(trajectory) > 1 else latest
    change = latest - prev

    leaderboard.append({
        "Startup": row["name"],
        "Momentum": latest,
        "Change": change
    })

leaderboard_df = pd.DataFrame(leaderboard)

# Sort by momentum (highest first)
leaderboard_df = leaderboard_df.sort_values(
    by="Momentum",
    ascending=False
).reset_index(drop=True)

# Create clean ranking numbers
leaderboard_df.index = leaderboard_df.index + 1
leaderboard_df.index.name = "Rank"

# Show Top 10
display(leaderboard_df.head(10))
```
